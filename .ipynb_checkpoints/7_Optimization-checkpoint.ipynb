# Optimization 최적화


---

## Generalization (일반화)

- Generalization(일반화)이란, 학습 데이터(Train data) 와 Input data(test data)가 달라져도, 출력에 대한 성능 차이가 나지 않게 하는 것을 의미합니다.

- 학습된 모델이 보이지 않는 데이터에서 얼마나 잘 작동할 것인지.

---

# Regularization (정규화)
Generalization (일반화)를 잘 하려고, Regularization (정규화) 작업

- Early stopping 



- Parameter norm penalty 
    - 부드러운 함수로 만들자
    - 모델의 가중치(weight) 파라미터에 대한 추가 제약을 부과하여 모델을 더 부드럽게 만들어주는 기술입니다., L2 norm L1 norm



- Data augmentation 
    - 데이터는 많을 수록 좋으니까
    - 데이터 확장,변형
    - 이미지 처리에서 데이터 증강은 이미지를 회전, 반전, 크기 조절, 색상 변환 등 다양한 방법으로 변형하여 훈련 데이터를 더 다양하게 만듭니다.
    



- Noise robustness
   
    - Add random noises inputs or weights.
    - 랜덤 노이즈 추가(Add Random Noises)
    - 데이터 정제(Data Augmentation): 훈련 데이터에 노이즈나 오류가 있는 경우
    
- Label smoothing 

    - 레이블 스무딩은 분류 작업에서 레이블에 대한 부분적인 불확실성을 고려하는 기법입니다.
    
    - 레이블에 대한 확신도를 조금 낮추어 모델이 레이블에 과도하게 확신하지 않도록 합니다.
    
    - Mix-up : 두 레이블 간의 혼합(Mix-up):
    - CutMix : 두 레이블 간 "cut and paste" 두 개의 무작위로 선택한 훈련 데이터 샘플을 고릅니다.
    - 하나의 이미지에서 임의로 선택한 영역을 잘라내고, 다른 이미지의 해당 영역으로 대체합니다. 이 과정을 "cut and paste"라고 합니다.
    
    
- Dropout
    - 일부 뉴런(neuron)을 무작위로 선택하여 출력을 제거하는 기술입니다.

    - 드롭아웃은 신경망에서 사용되는 정규화 기법 중 하나로, *훈련 중 무작위로 일부 뉴런을 비활성화*시킵니다.
    
    - 이를 통해 모델이 특정 뉴런에 지나치게 의존하지 않도록 하고, 일반화 능력을 향상시킵니다.

- Batch normalization
    - 신경망의 각 층(layer)에서 중간에서 정규화

    - 배치 정규화는 *신경망 내부의 중간 출력값을 정규화*하여 훈련을 안정화시키는 기법입니다.
    
    - 이를 통해 모델이 빠르게 수렴하고 과적합을 줄이는 데 도움이 됩니다.

## Over-fitting vs Under-fitting 


- 과적합(Overfitting)

    - 과적합은 모델이 훈련 데이터에 너무 맞추어져서 훈련 데이터에 대한 성능은 높지만 새로운 데이터나 테스트 데이터에 대한 일반화 능력이 떨어질 때 발생합니다.
    
    - 모델이 훈련 데이터의 노이즈나 특정 예외적인 패턴까지 학습하게 되면, 이러한 학습이 새로운 데이터에는 적합하지 않습니다.
    
    - 훈련 데이터에서는 성능이 우수하지만, 테스트 데이터에서는 성능이 저조하며 모델이 너무 복잡한 경우에 나타납니다.
       
    - 해결 방법: 모델의 복잡성을 줄이거나, 더 많은 훈련 데이터를 수집하여 일반화 능력을 향상시키는 등의 방법을 사용해야 합니다.
    

- 과소적합(Underfitting)

    - 과소적합은 모델이 너무 간단하거나 복잡하지 않아서 데이터의 복잡한 패턴이나 특징을 제대로 학습하지 못할 때 발생합니다.

    - 모델이 너무 단순한 경우, 실제 데이터의 다양한 패턴을 포착하지 못하고 오차가 크게 납니다.

    - 훈련 데이터와 테스트 데이터 모두에서 성능이 낮고, 모델이 데이터를 제대로 이해하지 못하는 경우에 나타납니다.

    - 해결 방법: 모델의 복잡성을 높이거나, 더 많은 특징을 추가하여 모델을 더 복잡하게 만들어야 합니다.


---
- 과소적합은 모델이 단순하여 데이터의 복잡성을 놓치는 문제를 나타내며, 

- 과적합은 모델이 훈련 데이터에 지나치게 맞춰져서 새로운 데이터에 대한 예측 능력이 부족한 문제를 나타냅니다. 

- 적절한 모델 복잡성과 데이터 양을 고려하여 이 두 문제를 균형 있게 다루는 것이 모델 개발에서 중요합니다.





## Cross validation 

- 모델을 학습할 때, 일반적으로 학습 데이터와 테스트 데이터로 나누어 사용합니다. 

    - 그러나 이런 방식은 데이터의 어떤 부분이 학습에 사용되고 어떤 부분이 테스트에 사용되는지에 따라 모델의 성능 측정에 영향을 미칠 수 있습니다.
    
- 교차 검증은 이러한 문제를 해결하기 위해 사용됩니다. 

- 데이터를 여러 개의 부분집합 또는 "폴드"로 나누고, 각각의 폴드를 사용하여 모델을 여러 번 훈련하고 평가합니다. 

- 예를 들어, 5-폴드 교차 검증은 데이터를 5개 부분으로 나누고 5번의 모델 훈련 및 평가를 수행합니다.

- 이렇게 하면 각 폴드가 번갈아가며 테스트 데이터로 사용되므로 모델이 데이터 전체에 대해 어떻게 일반화되는지 더 신뢰할 수 있는 평가를 얻을 수 있습니다. 

- 마지막으로, 이러한 평가 결과를 평균화하여 모델의 성능을 요약하거나 최적의 하이퍼파라미터를 선택하는 데 사용할 수 있습니다.

교차 검증은 모델의 일반화 능력을 평가하고, 과적합(Overfitting)을 탐지하는 데 도움이 되며, 모델의 신뢰성 있는 평가를 제공합니다. 따라서 머신 러닝 모델을 개발하거나 평가할 때 중요한 기술 중 하나입니다.

---

## Bias-variance tradeoff 

- Variance 분산, 모델 에서, 많이 퍼져 있는지, high
- Bias 편향, 모델 에서, 정답과 많이 떨어져 있는지 high

    - 편향(Bias)은 예측값이 정답과 얼마나 멀리 떨어져 있는지로 측정이 가능하고, 분산(Variance)은 예측값들끼리의 차이로 측정할 수 있으므로, 오류를 최소화하려면 편향과 분산의 합이 최소가 되는 지점을 찾는 것이 중요합니다. 
    
---

- We can derive that what we are minimizing (cost) can be decomposed into three different parts: bias2, variance, and noise.


- 노이즈가 있는 상태에서, bias, variance 를 잘 조절해야 함


- 편향(Bias): 모델의 편향은 모델이 실제 데이터의 복잡성을 적절하게 모델링하지 못할 때 나타납니다. 
    - 즉, 모델이 너무 단순하면 편향이 높다고 말합니다. 
    - 높은 편향은 모델이 데이터의 주요 패턴을 놓치고 예측 오차가 크게 될 수 있음을 의미합니다. 
    - 이를 "과소적합(Underfitting)"이라고도 합니다.

- 분산(Variance): 모델의 분산은 모델이 훈련 데이터에 과도하게 적합되어 있는 정도를 나타냅니다. 
    - 즉, 모델이 너무 복잡하면 분산이 높다고 말합니다. 
    - 높은 분산은 모델이 훈련 데이터에 너무 민감하게 반응하여 작은 변화에도 예측이 크게 흔들릴 수 있음을 의미합니다. 
    - 이를 "과적합(Overfitting)"이라고도 합니다.

- 잡음(Noise): 잡음은 데이터에서 발생하는 무작위한 변동을 나타냅니다. 
    - 모델을 개선할 수 없는 이러한 무작위한 변동이 존재합니다. 
    - 잡음은 모델이 어떤 정도의 오차를 가질 수밖에 없음을 의미합니다.

## 부트스트래핑(Bootstrapping)

   - 부트스트래핑은 무작위 샘플링과 재추출(복원)을 사용하는 테스트 또는 측정 기법입니다.
   
   - 즉, 데이터에서 무작위로 일부 샘플을 선택하고, 이 선택된 샘플을 다시 데이터에 넣고 다음 선택에서도 사용합니다.
   
   - 이것은 통계적인 추정이나 모델 평가에 사용됩니다. 
   
   - 예를 들어, 부트스트래핑을 사용하여 데이터셋에서 여러 샘플을 반복적으로 추출하여 신뢰 구간을 계산하거나 모델의 성능을 평가할 때 활용할 수 있습니다.

---

## 배깅(Bagging - Bootstrapping Aggregating)

   - 배깅은 부트스트래핑을 활용한 기계 학습 앙상블 기법 중 하나입니다.
   - 여러 개의 모델(기본 분류기)을 동시에 훈련하고, 각 모델은 부트스트랩 샘플(랜덤한 부분 데이터)에서 훈련됩니다.
   
   - 각 모델이 예측을 수행하면, 그 예측 결과를 집계하여 최종 예측을 만듭니다. 이 때 주로 투표 또는 평균을 사용합니다.
   
   - 배깅을 사용하면 모델의 분산을 줄이고, 예측 성능을 향상시키며, 과적합을 방지하는 데 도움이 됩니다.

---

## 부스팅(Boosting)

   - 부스팅은 앙상블 학습 기법 중 하나로, 여러 개의 약한 학습자(weak learner)를 순차적으로 학습시켜 강한 학습자(strong model)로 결합시키는 방식입니다.
   
   - 각 약한 학습자는 이전 학습자가 잘못 예측한 샘플에 더 큰 가중치를 주어 학습합니다. 이렇게 하면 어려운 샘플에 집중하여 모델을 개선합니다.
   - 부스팅은 순차적으로 학습하기 때문에 앞선 모델이 틀린 예측을 하면 이를 수정하는 방향으로 진행합니다.
   
   - 이를 통해 약한 학습자를 조합하여 높은 정확도를 갖는 강한 학습 모델을 생성할 수 있습니다.
   
   - 부스팅은 주로 분류 및 회귀 문제의 예측 성능 향상을 위한 목적으로 사용됩니다.
   

---

간단히 말하면, 배깅은 병렬적으로 모델을 훈련하고 집계하는 반면, 부스팅은 순차적으로 약한 모델을 학습하여 예측 성능을 향상시키는 앙상블 방법입니다.

---

## Ensemble 기법

- Ensemble 기법은 여러 개의 기계 학습 모델을 결합하여 더 강력하고 정확한 모델을 생성하는 기술입니다. 

- 모델의 예측 성능을 향상시키고, 과적합을 줄이는 데 도움이 됩니다. 

- 이러한 기법을 사용하면 개별 모델의 약점을 보완하고 예측 성능을 향상시킬 수 있습니다. 
    - Bagging - 여러개 동시에 병렬
    - Boosting - 약한 학습자 모델 순차적 강화 강한 학습자로 
    - Voting
    - Stacking - 여러 개의 다른 모델을 조합하여 하나의 메타 모델(meta-model)을 만드는 방식
